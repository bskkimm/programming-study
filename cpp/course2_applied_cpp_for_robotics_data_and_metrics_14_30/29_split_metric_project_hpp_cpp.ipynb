## Concept: real C++ file structure
Notebooks are for learning. Real C++ projects split public declarations and implementation.

## Header file: `mini_metrics.hpp`
The header declares structs and functions.

````c++17
// mini_metrics.hpp idea
// #ifndef MINI_METRICS_HPP_
// #define MINI_METRICS_HPP_
// #include <string>
// struct MetricResult { double score{0.0}; bool available{false}; std::string reason{"unavailable"}; };
// MetricResult calculate_speed_limit_compliance(double speed_mps, double speed_limit_mps);
// #endif
````

## Source file: `mini_metrics.cpp`
The source file defines function bodies.

````c++17
// mini_metrics.cpp idea
// #include "mini_metrics.hpp"
// MetricResult calculate_speed_limit_compliance(double speed_mps, double speed_limit_mps)
// {
//   MetricResult result;
//   result.available = true;
//   result.score = 1.0;
//   result.reason = "available";
//   if (speed_mps > speed_limit_mps) { result.score = 0.0; result.reason = "speed_too_high"; }
//   return result;
// }
````

## Main file: `main.cpp`
The application includes the header and uses the function.

````c++17
// main.cpp idea
// #include <iostream>
// #include "mini_metrics.hpp"
// int main(){ auto result = calculate_speed_limit_compliance(12.0, 10.0); std::cout << result.reason << std::endl; }
````
