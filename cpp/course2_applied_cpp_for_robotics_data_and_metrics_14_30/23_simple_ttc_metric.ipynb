## Concept: TTC
TTC means time-to-collision. Sometimes it is unavailable, so we use `std::optional<double>`.

```c++
#include <iostream>
#include <string>
#include <optional>
```

```c++
struct MetricResult23 { 
    double score{0.0}; 
    bool available{false}; 
    std::string reason{"unavailable"}; 
    double ttc_s{-1.0}; };
```

## Implement TTC metric

```c++
auto ttc_metric23 = [](std::optional<double> ttc_s, double threshold_s) -> MetricResult23
{
  MetricResult23 result;
  if (!ttc_s.has_value()) {
    result.reason = "unavailable_no_ttc";
    return result;
  }

  result.available = true;
  result.ttc_s = ttc_s.value();
  result.score = 1.0;
  result.reason = "available";

  if (ttc_s.value() < threshold_s) {
    result.score = 0.0;
    result.reason = "ttc_too_low";
  }
  return result;
};

{
  const auto fail = ttc_metric23(1.5, 2.0);
  const auto pass = ttc_metric23(3.0, 2.0);
  const auto fail_or_pass = ttc_metric23(1.5, 1.0);
  const auto unavailable = ttc_metric23(std::nullopt, 2.0);

  std::cout << fail.reason << std::endl;
  std::cout << pass.reason << std::endl;
  std::cout << "Available?:" << fail_or_pass.reason << ", Score:" << fail_or_pass.score << std::endl;
  std::cout << unavailable.reason << std::endl;
}
```
