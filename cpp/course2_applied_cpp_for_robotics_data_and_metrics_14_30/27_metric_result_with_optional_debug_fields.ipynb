## Concept: debug fields
Scores are not enough. You need raw value and infraction time for debugging.

```c++
#include <iostream>
#include <string>
#include <optional>
```

```c++
struct MetricResult27
{
  double score{0.0};
  bool available{false};
  std::string reason{"unavailable"};
  std::optional<double> raw_value{std::nullopt};
  std::optional<double> infraction_time_s{std::nullopt};
};
```

```c++
auto print_result27 = [](const MetricResult27 & r) -> void
{
  std::cout << std::boolalpha << "available = " << r.available << ", score = " << r.score << ", reason = " << r.reason << std::endl;
  if (r.raw_value.has_value()) std::cout << "raw_value = " << r.raw_value.value() << std::endl;
  if (r.infraction_time_s.has_value()) std::cout << "infraction_time_s = " << r.infraction_time_s.value() << std::endl;
};

auto speed_metric27 = [](double speed_mps, double limit_mps, double time_s) -> MetricResult27
{
  MetricResult27 result;
  result.available = true;
  result.score = 1.0;
  result.reason = "available";
  result.raw_value = speed_mps;
  if (speed_mps > limit_mps) {
    result.score = 0.0;
    result.reason = "speed_too_high";
    result.infraction_time_s = time_s;
  }
  return result;
};

{
  print_result27(speed_metric27(12.0, 10.0, 3.5));
  print_result27(speed_metric27(8.0, 10.0, 3.5));
}
```
