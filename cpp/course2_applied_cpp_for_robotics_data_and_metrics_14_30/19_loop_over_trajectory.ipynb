## Range-based loop
Use this when you do not need the index.

```c++17
struct TrajectoryPoint19
{
  double x_m;
  double y_m;
  double speed_mps;
  double time_s;
};

auto make_sample_trajectory19 = []() -> std::vector<TrajectoryPoint19>
{
  return {
    {0.0, 0.0, 2.0, 0.0},
    {1.0, 0.0, 2.5, 0.5},
    {2.0, 0.0, 3.0, 1.0},
    {3.0, 0.0, 3.5, 1.5}
  };
};
{
  const auto trajectory = make_sample_trajectory19();
  for (const auto & point : trajectory) {
    std::cout << "time = " << point.time_s << ", speed = " << point.speed_mps << std::endl;
  }
}
```

## Index-based loop
Use this when you need `i`, previous point, or next point.

```c++17
{
  const auto trajectory = make_sample_trajectory19();
  for (std::size_t i = 0; i < trajectory.size(); ++i) { //for (initialize; condition; update)
    std::cout << "index = " << i << ", x = " << trajectory.at(i).x_m << std::endl;
  }
}
```

## Violation detection
Break early when a failure is found.

```c++17
{
  const auto trajectory = make_sample_trajectory19();
  const double limit = 3.0;
  bool violation = false;
  for (const auto & point : trajectory) {
    if (point.speed_mps > limit) {
      violation = true;
      std::cout << "violation at t = " << point.time_s << std::endl;
      break;
    }
  }
  std::cout << std::boolalpha << violation << std::endl;
}
```

## Exercise
Print only points whose speed is greater than `2.8`.

```c++17
// Exercise starter
{
  const auto trajectory = make_sample_trajectory19();
  // TODO: loop and print speed > 2.8
  for (std::size_t i = 0; i < trajectory.size(); ++i) {
    if (trajectory.at(i).speed_mps > 2.8) {
      std::cout << "speed violation at t = " << trajectory.at(i).time_s << std::endl;
    }
  }
}
```

## Solution

```c++17
{
  const auto trajectory = make_sample_trajectory19();
  for (const auto & point : trajectory) {
    if (point.speed_mps > 2.8) {
      std::cout << point.time_s << ": " << point.speed_mps << std::endl;
    }
  }
}
```
