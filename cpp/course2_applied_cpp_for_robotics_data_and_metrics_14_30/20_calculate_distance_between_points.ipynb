## Concept: distance
Distance is used in collision checks, progress, map proximity, and object filtering.

```c++17
#include <iostream>
#include <cmath>
```

```c++17
struct Point2D20 { double x_m{0.0}; double y_m{0.0}; };
```

## Distance helper
`std::hypot(dx, dy)` computes `sqrt(dx*dx + dy*dy)`.

```c++17
double distance20(const Point2D20 & a, const Point2D20 & b)
{
  return std::hypot(b.x_m - a.x_m, b.y_m - a.y_m);
}
```

```c++17
{
  Point2D20 a{0,0};
  Point2D20 b{3,4};
  std::cout << distance20(a,b) << std::endl;
}
```

## Metric-like threshold check

```c++17
bool is_too_close20(const Point2D20 & ego, const Point2D20 & object, double threshold_m)
{
  return distance20(ego, object) < threshold_m;
}

{
  Point2D20 ego{0,0};
  Point2D20 object{1,0.5};
  std::cout << std::boolalpha << is_too_close20(ego, object, 2.0) << std::endl;
}
```
