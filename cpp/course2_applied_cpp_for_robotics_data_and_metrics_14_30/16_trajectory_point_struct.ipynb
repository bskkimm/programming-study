## Concept: trajectory point
A trajectory point is a vehicle state at a future time. It usually has pose, speed, and time.

```c++17
#include <iostream>
```

```c++17
struct Pose2D16
{
  double x_m{0.0};
  double y_m{0.0};
  double yaw_rad{0.0};
};

struct TrajectoryPoint16
{
  Pose2D16 pose;
  double speed_mps{0.0};
  double time_s{0.0};
};
```

## Nested field access
`point.pose.x_m` means: point → pose → x.

```c++17
{
  TrajectoryPoint16 point{{5.0, 1.0, 0.0}, 3.5, 0.5};
  std::cout << "x = " << point.pose.x_m << std::endl;
  std::cout << "speed = " << point.speed_mps << std::endl;
  std::cout << "time = " << point.time_s << std::endl;
}
```

## Helper: stopped check
Metrics often check low-speed or stopped ego state.

```c++17
bool is_stopped16(const TrajectoryPoint16 & point, double threshold_mps)
{
  return point.speed_mps < threshold_mps;
}

{
  TrajectoryPoint16 point{{0.0, 0.0, 0.0}, 0.01, 0.0};
  std::cout << std::boolalpha << is_stopped16(point, 0.05) << std::endl;
}
```
