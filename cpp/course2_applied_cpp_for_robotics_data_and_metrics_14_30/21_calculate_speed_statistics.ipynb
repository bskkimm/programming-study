## Concept: speed statistics
Metrics often need average speed, max speed, and speed violations.

```c++17
#include <iostream>
#include <vector>
```

## Average speed

```c++17
struct SpeedPoint21 { double speed_mps; double time_s; };
auto make_speed_data21 = []() -> std::vector<SpeedPoint21> { return {{2,0}, {3.5,0.5}, {4,1}, {2.5,1.5}}; };

auto average_speed21 = [](const std::vector<SpeedPoint21> & points) -> double
{
  if (points.empty()) return 0.0;
  double sum = 0.0;
  for (const auto & p : points) sum += p.speed_mps;
  return sum / points.size();
};

{
  std::cout << average_speed21(make_speed_data21()) << std::endl;
}
```

## Maximum speed

```c++17
auto max_speed21 = [](const std::vector<SpeedPoint21> & points) -> double
{
  if (points.empty()) return 0.0;
  double max_speed = points.front().speed_mps;
  for (const auto & p : points) {
    if (p.speed_mps > max_speed) max_speed = p.speed_mps;
  }
  return max_speed;
};

{
  std::cout << max_speed21(make_speed_data21()) << std::endl;
}
```

## Exercise
Count stopped points with speed `< 0.05`.

```c++17
// Exercise starter
auto count_stopped21 =[](const std::vector<SpeedPoint21> & points) -> int
{
  // TODO
  if (points.empty()) return 0;
  int count = 0;
  for (const auto & p : points)
      if (p.speed_mps < 0.05) count++;

  return count;
};

{
  std::cout << count_stopped21(make_speed_data21()) << std::endl;
}
```

## Solution

```c++17
auto count_stopped21_solution = [](const std::vector<SpeedPoint21> & points) -> int
{
  int count = 0;
  for (const auto & p : points) {
    if (p.speed_mps < 0.05) ++count;
  }
  return count;
};

{
  std::vector<SpeedPoint21> points{{0,0}, {0.01,0.5}, {2,1}};
  std::cout << count_stopped21_solution(points) << std::endl;
}
```
