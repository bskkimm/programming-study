## Concept: ahead and behind
Assume ego faces positive x. Then larger object x means ahead; smaller x means behind.

```c++
#include <iostream>
#include <cmath>
```

```c++
struct Pose2D26 { double x_m; double y_m; double yaw_rad; };
struct ObjectState26 { double x_m; double y_m; };
```

```c++
auto is_ahead26 = [](const Pose2D26 & ego, const ObjectState26 & object) -> bool 
{ return object.x_m > ego.x_m; };
auto is_behind26 = [](const Pose2D26 & ego, const ObjectState26 & object) -> bool 
{ return object.x_m < ego.x_m; };
auto is_lateral26 = [](const Pose2D26 & ego, const ObjectState26 & object) -> bool 
{ return std::abs(object.y_m - ego.y_m) > 2.0; };

{
  Pose2D26 ego{0,0,0};
  ObjectState26 object{5,1};
  std::cout << std::boolalpha;
  std::cout << "ahead = " << is_ahead26(ego, object) << std::endl;
  std::cout << "behind = " << is_behind26(ego, object) << std::endl;
  std::cout << "lateral = " << is_lateral26(ego, object) << std::endl;
}
```

## Robotics connection
Real Autoware-style logic uses yaw and relative angle, but this simplified version teaches the classification idea.
