## Concept: aggregation
Autonomous driving evaluation combines many metrics. Some are hard gates; others are averaged or weighted.

```c++17
#include <iostream>
#include <string>
#include <vector>
```

```c++17
struct MetricResult28 { double score{0.0}; bool available{false}; std::string reason{"unavailable"}; };
```

## Hard gate aggregation
If any child is unavailable, final is unavailable. If any child fails, final fails.

```c++17
MetricResult28 hard_gate28(const std::vector<MetricResult28> & metrics)
{
  MetricResult28 result;
  if (metrics.empty()) { result.reason = "unavailable_no_metrics"; return result; }
  result.available = true; result.score = 1.0; result.reason = "available";
  for (const auto & m : metrics) {
    if (!m.available) { result.available = false; result.score = 0.0; result.reason = "unavailable_child_metric"; return result; }
    if (m.score <= 0.0) { result.score = 0.0; result.reason = "failed_child_metric"; return result; }
  }
  return result;
}
```

```c++17
{
  MetricResult28 speed{1.0, true, "available"};
  MetricResult28 lane{0.0, true, "lane_deviation_too_large"};
  const auto result = hard_gate28({speed, lane});
  std::cout << result.reason << ", score = " << result.score << std::endl;
}
```

## Weighted average

```c++17
double weighted_average28(double a, double wa, double b, double wb)
{
  return (a * wa + b * wb) / (wa + wb);
}

{
  std::cout << weighted_average28(1.0, 5.0, 0.5, 2.0) << std::endl;
}
```
