## Concept: trajectory as vector
A trajectory is a sequence of future points, so `std::vector<TrajectoryPoint>` is natural.

```c++17
#include <iostream>
#include <vector>
```

```c++17
struct TrajectoryPoint18
{
  double x_m{0.0};
  double y_m{0.0};
  double speed_mps{0.0};
  double time_s{0.0};
};
```

## Build a trajectory

```c++17
{
  std::vector<TrajectoryPoint18> trajectory;
  trajectory.push_back({0.0, 0.0, 2.0, 0.0});
  trajectory.push_back({1.0, 0.0, 2.0, 0.5});
  trajectory.push_back({2.0, 0.0, 2.0, 1.0});

  std::cout << "size = " << trajectory.size() << std::endl;
  std::cout << "first x = " << trajectory.front().x_m << std::endl;
  std::cout << "last x = " << trajectory.back().x_m << std::endl;
  std::cout << "middle x with index" << trajectory[1].x_m << std::endl;
}
```

## Safety: check empty first
Do not call `front()` or `back()` on an empty vector.

```c++17
{
  std::vector<TrajectoryPoint18> trajectory;
  if (trajectory.empty()) {
    std::cout << "empty trajectory" << std::endl;
  }
}
```

## Reserve capacity
Useful when you know approximate trajectory size.

```c++17
{
  std::vector<TrajectoryPoint18> trajectory;
  trajectory.reserve(100);
  std::cout << "capacity = " << trajectory.capacity() << std::endl;
}
```
