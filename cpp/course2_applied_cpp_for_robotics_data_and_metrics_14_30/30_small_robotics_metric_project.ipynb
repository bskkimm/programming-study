## Capstone goal
Build a tiny pure-C++ metric system: speed metric + progress metric + aggregation.

```c++17
#include <iostream>
#include <string>
#include <vector>
#include <cmath>
```

```c++17
struct MetricResult30 { double score{0.0}; bool available{false}; std::string reason{"unavailable"}; };
struct TrajectoryPoint30 { double x_m{0.0}; double y_m{0.0}; double speed_mps{0.0}; double time_s{0.0}; };
```

## Speed limit metric

```c++17
MetricResult30 speed_limit_metric30(const std::vector<TrajectoryPoint30> & trajectory, double speed_limit_mps)
{
  MetricResult30 result;
  if (trajectory.empty()) { result.reason = "unavailable_empty_trajectory"; return result; }
  result.available = true; result.score = 1.0; result.reason = "available";
  for (const auto & p : trajectory) {
    if (p.speed_mps > speed_limit_mps) { result.score = 0.0; result.reason = "speed_too_high"; return result; }
  }
  return result;
}
```

## Progress metric

```c++17
double distance30(const TrajectoryPoint30 & a, const TrajectoryPoint30 & b)
{
  return std::hypot(b.x_m - a.x_m, b.y_m - a.y_m);
}

MetricResult30 progress_metric30(const std::vector<TrajectoryPoint30> & trajectory, double minimum_progress_m)
{
  MetricResult30 result;
  if (trajectory.size() < 2) { result.reason = "unavailable_short_trajectory"; return result; }
  result.available = true; result.score = 1.0; result.reason = "available";
  if (distance30(trajectory.front(), trajectory.back()) < minimum_progress_m) {
    result.score = 0.0; result.reason = "insufficient_progress";
  }
  return result;
}
```

## Aggregation

```c++17
MetricResult30 aggregate30(const MetricResult30 & a, const MetricResult30 & b)
{
  MetricResult30 result;
  if (!a.available || !b.available) { result.reason = "unavailable_child_metric"; return result; }
  result.available = true;
  result.score = a.score * b.score;
  result.reason = result.score > 0.0 ? "available" : "failed_child_metric";
  return result;
}
```

## Run all metrics

```c++17
{
  std::vector<TrajectoryPoint30> trajectory{{0,0,2,0}, {2,0,3,0.5}, {5,0,4,1.0}};
  const auto speed = speed_limit_metric30(trajectory, 10.0);
  const auto progress = progress_metric30(trajectory, 3.0);
  const auto final_result = aggregate30(speed, progress);

  std::cout << "speed = " << speed.reason << std::endl;
  std::cout << "progress = " << progress.reason << std::endl;
  std::cout << "final = " << final_result.reason << ", score = " << final_result.score << std::endl;
}
```
