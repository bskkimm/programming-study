## Concept: pose
A simplified robotics pose contains `x`, `y`, and `yaw`. We add units in names: `_m` for meters, `_rad` for radians.

```c++17
#include <iostream>
#include <cmath>
```

```c++17
struct Pose2D15
{
  double x_m{0.0};
  double y_m{0.0};
  double yaw_rad{0.0};
};
```

## Create pose objects

```c++17
{
  Pose2D15 ego{10.0, 2.0, 0.0};
  std::cout << "x = " << ego.x_m << std::endl;
  std::cout << "y = " << ego.y_m << std::endl;
  std::cout << "yaw = " << ego.yaw_rad << std::endl;
}
```

## Helper function using const reference
`const Pose2D15 &` means read the pose without copying or modifying it.

```c++17
void print_pose15(const Pose2D15 & pose)
{
  std::cout << "x = " << pose.x_m
            << ", y = " << pose.y_m
            << ", yaw = " << pose.yaw_rad << std::endl;
}

{
  Pose2D15 start{0.0, 0.0, 0.0};
  print_pose15(start);
}
```

## Robotics connection
