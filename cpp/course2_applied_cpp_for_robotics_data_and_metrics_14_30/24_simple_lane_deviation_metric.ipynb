## Concept: lane deviation
Lane keeping can be simplified as: fail if lateral deviation is too large.

```c++17
#include <iostream>
#include <string>
#include <optional>
```

```c++17
struct MetricResult24 { double score{0.0}; bool available{false}; std::string reason{"unavailable"}; double lateral_deviation_m{-1.0}; };
```

```c++17
MetricResult24 lane_deviation_metric24(std::optional<double> deviation_m, double max_deviation_m)
{
  MetricResult24 result;
  if (!deviation_m.has_value()) {
    result.reason = "unavailable_no_lane_info";
    return result;
  }

  result.available = true;
  result.lateral_deviation_m = deviation_m.value();
  result.score = 1.0;
  result.reason = "available";

  if (deviation_m.value() > max_deviation_m) {
    result.score = 0.0;
    result.reason = "lane_deviation_too_large";
  }
  return result;
}
```

```c++17
{
  std::cout << lane_deviation_metric24(0.2, 0.5).reason << std::endl;
  std::cout << lane_deviation_metric24(0.8, 0.5).reason << std::endl;
  std::cout << lane_deviation_metric24(std::nullopt, 0.5).reason << std::endl;
}
```

## Boundary discussion
At exactly `0.5`, this passes because the violation operator is `>`, not `>=`.

```c++17
{
  const auto boundary = lane_deviation_metric24(0.5, 0.5);
  std::cout << boundary.reason << std::endl;
}
```
