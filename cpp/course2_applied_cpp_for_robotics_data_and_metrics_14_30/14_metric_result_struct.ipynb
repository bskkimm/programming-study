## Concept: why metric results need a struct
A robotics metric usually returns multiple things: `score`, `available`, and `reason`. Keeping them in one `struct` prevents inconsistent states.

```c++17
#include <iostream>
#include <string>
```

## Define the result type
Default values mean a new result starts as unavailable.

```c++17
struct MetricResult14
{
  double score{0.0};
  bool available{false};
  std::string reason{"unavailable"};
};
```

## Create and inspect a default result

```c++17
{
  MetricResult14 result;
  std::cout << std::boolalpha;
  std::cout << "score = " << result.score << std::endl;
  std::cout << "available = " << result.available << std::endl;
  std::cout << "reason = " << result.reason << std::endl;
}
```

## Function returning a complete metric result
This is the core pattern you will reuse constantly.

```c++17
MetricResult14 calculate_dummy_metric14(bool input_available)
{
  MetricResult14 result;
  if (!input_available) {
    result.reason = "unavailable_missing_input";
    return result;
  }

  result.available = true;
  result.score = 1.0;
  result.reason = "available";
  return result;
}
```

```c++17
{
  const auto pass = calculate_dummy_metric14(true);
  const auto unavailable = calculate_dummy_metric14(false);

  std::cout << pass.reason << ", score = " << pass.score << std::endl;
  std::cout << unavailable.reason << ", score = " << unavailable.score << std::endl;
}
```
