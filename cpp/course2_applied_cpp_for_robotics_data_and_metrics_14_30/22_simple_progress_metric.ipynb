## Concept: progress metric
Progress can be approximated as distance from first point to last point.

```c++17
#include <iostream>
#include <string>
#include <vector>
#include <cmath>
```

```c++17
struct MetricResult22 { double score{0.0}; bool available{false}; std::string reason{"unavailable"}; double raw_progress_m{0.0}; };
struct Point2D22 { double x_m; double y_m; };
struct TrajectoryPoint22 { Point2D22 position; double time_s; };

double distance22(const Point2D22 & a, const Point2D22 & b)
{
  return std::hypot(b.x_m - a.x_m, b.y_m - a.y_m);
}
```

## Implement metric
Unavailable if fewer than two points. Fail if progress is below threshold.

```c++17
MetricResult22 progress_metric22(const std::vector<TrajectoryPoint22> & trajectory, double min_progress_m)
{
  MetricResult22 result;
  if (trajectory.size() < 2) {
    result.reason = "unavailable_short_trajectory";
    return result;
  }

  result.raw_progress_m = distance22(trajectory.front().position, trajectory.back().position);
  result.available = true;
  result.score = 1.0;
  result.reason = "available";

  if (result.raw_progress_m < min_progress_m) {
    result.score = 0.0;
    result.reason = "insufficient_progress";
  }
  return result;
}
```

```c++17
{
  std::vector<TrajectoryPoint22> trajectory{{{0,0},0}, {{10,0},1}};
  const auto result = progress_metric22(trajectory, 5.0);
  std::cout << result.reason << ", progress = " << result.raw_progress_m << std::endl;
}
```
