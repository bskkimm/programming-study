## Concept: object state
Perception objects have position, speed, and classification such as `car`, `pedestrian`, or `unknown`.

```c++17
#include <iostream>
#include <string>
```

```c++17
struct Pose2D17 { double x_m{0.0}; double y_m{0.0}; double yaw_rad{0.0}; };

struct ObjectState17
{
  Pose2D17 pose;
  double speed_mps{0.0};
  std::string classification{"unknown"};
};
```

## Create object state

```c++17
{
  ObjectState17 object{{12.0, 1.0, 0.0}, 0.0, "car"};
  std::cout << object.classification << " at x = " << object.pose.x_m << std::endl;
}
```

## Unknown object check
Some metrics ignore unknown objects to avoid unstable tracking artifacts.

```c++17
bool is_unknown_object17(const ObjectState17 & object)
{
  return object.classification == "unknown";
}

{
  ObjectState17 object{{1.0, 0.0, 0.0}, 0.0, "unknown"};
  std::cout << std::boolalpha << is_unknown_object17(object) << std::endl;
}
```
