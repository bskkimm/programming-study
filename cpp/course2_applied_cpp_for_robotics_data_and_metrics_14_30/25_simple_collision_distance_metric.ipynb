## Concept: collision distance
Before polygon collision, learn simple point-to-point distance thresholding.

```c++17
#include <iostream>
#include <string>
#include <cmath>
```

```c++17
struct MetricResult25 { double score{0.0}; bool available{false}; std::string reason{"unavailable"}; double distance_m{-1.0}; };
struct Point2D25 { double x_m; double y_m; };

double distance25(const Point2D25 & a, const Point2D25 & b){ return std::hypot(b.x_m-a.x_m, b.y_m-a.y_m); }
```

```c++17
MetricResult25 collision_distance_metric25(const Point2D25 & ego, const Point2D25 & object, double collision_distance_m)
{
  MetricResult25 result;
  result.available = true;
  result.score = 1.0;
  result.reason = "available";
  result.distance_m = distance25(ego, object);

  if (result.distance_m < collision_distance_m) {
    result.score = 0.0;
    result.reason = "object_too_close";
  }
  return result;
}
```

```c++17
{
  Point2D25 ego{0,0};
  Point2D25 object{1,0};
  const auto result = collision_distance_metric25(ego, object, 2.0);
  std::cout << result.reason << ", distance = " << result.distance_m << std::endl;
}
```
