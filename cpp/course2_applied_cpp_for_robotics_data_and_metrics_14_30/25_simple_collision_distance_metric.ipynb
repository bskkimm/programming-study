## Concept: collision distance
Before polygon collision, learn simple point-to-point distance thresholding.

```c++
#include <iostream>
#include <string>
#include <cmath>
```

```c++
struct MetricResult25 { double score{0.0}; 
    bool available{false}; 
    std::string reason{"unavailable"}; 
    double distance_m{-1.0}; };
    
struct Point2D25 { double x_m; double y_m; };
```

```c++
auto distance25 = [](const Point2D25 & a, const Point2D25 & b) -> double { 
  return std::hypot(b.x_m-a.x_m, b.y_m-a.y_m); };

auto collision_distance_metric25 = [&](const Point2D25 & ego, const Point2D25 & object, double collision_distance_m) -> MetricResult25
{
  MetricResult25 result;
  result.available = true;
  result.score = 1.0;
  result.reason = "available";
  result.distance_m = distance25(ego, object);

  if (result.distance_m < collision_distance_m) {
    result.score = 0.0;
    result.reason = "object_too_close";
  }
  return result;
};

{
  Point2D25 ego{0,0};
  Point2D25 object{1,0}; Point2D25 object2{5,0};
  const auto result = collision_distance_metric25(ego, object, 2.0);
  const auto result2 = collision_distance_metric25(ego, object2, 2.0);
  std::cout << "Result 1: " << result.reason << ", distance = " << result.distance_m << std::endl;
  std::cout << "Result 2: " << result2.reason << ", distance = " << result2.distance_m << std::endl;
}
```
