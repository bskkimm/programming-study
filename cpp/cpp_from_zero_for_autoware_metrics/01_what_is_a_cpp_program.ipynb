## 1. C++ does not run directly

In Python, you usually do:

```bash
python main.py
```

Python reads your code and runs it.

C++ is different. C++ code must be **compiled** first.

```text
C++ source code
↓
compiler
↓
executable program
↓
run
```

In a terminal, the classic flow is:

```bash
g++ main.cpp -o main
./main
```

Meaning:

```text
main.cpp = your C++ source code
g++ = compiler
-o main = output executable name is main
./main = run the executable
```

In this notebook, we are using a C++ Jupyter kernel, so we can run C++ cells directly. But in real Autoware development, you still need the compile/build mental model.

## 2. Your first C++ program

A normal terminal-style C++ program starts with `main()`:

```cpp
#include <iostream>

int main()
{
  std::cout << "Hello Autoware" << std::endl;
  return 0;
}
```

In this notebook, we do not need to define `main()` in every cell. We can run the core statement directly.

````c++
#include <iostream>
#include <string>

const std::string first_message = "Hi, Autoware!";

std::cout << first_message << std::endl;
std::cout << "Hello Autoware" << std::endl;
````

Expected output:

```text
Hello Autoware
```

Tiny victory, but this is the root of everything.

## 3. Explain every line

### `#include <iostream>`

This means:

> I want to use input/output functions.

Specifically, we need it for:

```cpp
std::cout
```

Without this include, C++ does not know what `std::cout` is.

Think of `#include` like importing a tool. Python comparison:

```python
import something
```

C++:

```cpp
#include <something>
```

Not exactly the same, but close enough for now.

````c++
#include <iostream>

std::cout << "iostream lets us print text." << std::endl;
````

### `int main()`

Every normal C++ program starts from `main()`.

When you run the program, C++ looks for:

```cpp
main()
```

and begins there.

```cpp
int
```

means this function returns an integer.

So:

```cpp
int main()
```

means:

> This is the main function, and it returns an integer status code.

### `{}` curly braces

Curly braces define a block.

```cpp
int main()
{
  // inside main
}
```

Python uses indentation:

```python
def main():
    print("Hello")
```

C++ uses braces:

```cpp
int main()
{
  std::cout << "Hello" << std::endl;
}
```

### `std::cout << "Hello Autoware" << std::endl;`

This prints text.

Break it down:

```cpp
std::cout
```

means console output.

```cpp
<<
```

means send this to output.

```cpp
"Hello Autoware"
```

is text.

```cpp
std::endl
```

means end the line.

So:

```cpp
std::cout << "Hello Autoware" << std::endl;
```

means:

> Print `Hello Autoware` and go to the next line.

Python equivalent:

```python
print("Hello Autoware")
```

C++ is more verbose. Very C++ behavior.

````c++
#include <iostream>

std::cout << "First line" << std::endl;
std::cout << "Second line" << std::endl;
````

### `return 0;`

Inside `main()`, this means:

> The program finished successfully.

Usually:

```text
return 0 = success
return non-zero = error
```

For now, always write:

```cpp
return 0;
```

inside `main()` when writing terminal C++ programs.

## 4. Semicolon rule

In C++, most statements end with:

```cpp
;
```

Example:

```cpp
std::cout << "Hello" << std::endl;
return 0;
```

If you forget semicolon, C++ gets angry.

Python is chill. C++ is not chill.

## 6. Today’s vocabulary

| Word | Meaning |
|---|---|
| source code | Human-written code |

| `.cpp` | C++ source file |

| `{}` | Code block |
| `std::cout` | Console output |
