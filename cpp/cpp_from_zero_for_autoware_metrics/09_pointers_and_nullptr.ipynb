# Lecture 9 — Pointers and `nullptr`

Today’s goal:

> Understand the beginner-level meaning of pointers, `nullptr`, `*`, `&`, and `->`.

Do not panic. We only need the practical Autoware level first.

## 1. Address and pointer

A pointer stores the address of an object.

```cpp
int value = 10;
int * ptr = &value;
```

Meaning:

```text
value is an int.
ptr points to value.
&value means address of value.
```

````c++
#include <iostream>

{
  int value = 10;
  int * ptr = &value;

  std::cout << "value = " << value << std::endl;
  std::cout << "*ptr = " << *ptr << std::endl;
  std::cout << "ptr = " << ptr << std::endl;
}
````

`*ptr` means:

> Access the object that the pointer points to.

This is called dereferencing.

## 2. `nullptr`

A pointer may point to nothing.

```cpp
int * ptr = nullptr;
```

This is common in Autoware when optional context is missing.

````c++
#include <iostream>

{
  int * ptr = nullptr;

  if (ptr == nullptr) {
    std::cout << "ptr points to nothing" << std::endl;
  }
}
````

C++ style often writes:

```cpp
if (!ptr) {
  // ptr is null
}
```

That means the same as:

```cpp
if (ptr == nullptr) {
}
```

````c++
#include <iostream>

{
  int * ptr = nullptr;

  if (!ptr) {
    std::cout << "missing pointer" << std::endl;
  }
}
````

## 3. Dot `.` vs arrow `->`

If you have a normal object:

```cpp
result.score
```

If you have a pointer:

```cpp
result_ptr->score
```

`ptr->field` is shorthand for:

```cpp
(*ptr).field
```

````c++
#include <iostream>
#include <string>

struct MetricResult09
{
  double score{0.0};
  std::string reason{"unavailable"};
};

{
  MetricResult09 result;
  result.score = 1.0;
  result.reason = "available";

  MetricResult09 * result_ptr = &result;

  std::cout << "normal object: " << result.score << std::endl;
  std::cout << "through pointer: " << result_ptr->score << std::endl;
}
````

## 4. Reference vs pointer mental model

```text
reference = must exist
pointer = may be missing
nullptr = missing
```

This is not the full C++ truth, but it is a useful beginner rule.

## 5. Metric-style pointer input

Sometimes a helper may receive optional data as pointer.

If it is null, skip or return unavailable.

````c++
#include <iostream>
#include <vector>

int get_count_if_available_09(const std::vector<double> * values)
{
  if (!values) {
    return 0;
  }

  return static_cast<int>(values->size());
}

{
  std::vector<double> values{1.0, 2.0, 3.0};

  std::cout << "count with data = " << get_count_if_available_09(&values) << std::endl;
  std::cout << "count without data = " << get_count_if_available_09(nullptr) << std::endl;
}
````

## 6. Autoware-style thought

You may later see code like:

```cpp
if (!route_handler) {
  result.reason = "unavailable_no_route_handler";
  return result;
}
```

Meaning:

```text
If the route handler pointer is missing, this metric cannot be computed.
```

Write a function that receives a pointer to a double threshold. If the pointer is null, return default threshold `10.0`. Otherwise return the pointed value.

````c++
#include <iostream>

double get_threshold_09(const double * threshold_ptr)
{
  // TODO: if threshold_ptr is null, return 10.0.
  if (!threshold_ptr) {
    return 10.0;
  }
  // Otherwise return the value it points to.
  return *threshold_ptr;
}

{
  double custom_threshold = 5.0;

  std::cout << "custom = " << get_threshold_09(&custom_threshold) << std::endl;
  std::cout << "default = " << get_threshold_09(nullptr) << std::endl;
}
````
