# Lecture 2 — Variables and Types

Today’s goal:

> You understand how C++ stores values like score, speed, availability, and reason.

This is very important for Autoware metrics because almost every metric result contains values like:

```cpp
double score;
bool available;
std::string reason;
```

## 1. What is a variable?

A variable is a named box that stores a value.

Example:

```cpp
double score = 1.0;
```

Meaning:

```text
Create a variable named score.
Its type is double.
Its value is 1.0.
```

In Python, you can write:

```python
score = 1.0
```

But in C++, you must say the type:

```cpp
double score = 1.0;
```

C++ wants to know what kind of value this is: number, decimal, text, true/false, etc.

````c++
#include <iostream>

{
  double score = 1.0;
  std::cout << "score = " << score << std::endl;
}
````

## 2. Four basic types you need first

For metrics work, start with these four:

```cpp
int
double
bool
std::string
```

| Type | Meaning | Example |
|---|---|---|
| `int` | Integer | `3`, `10`, `-5` |
| `double` | Decimal number | `1.0`, `3.14`, `5.5` |
| `bool` | True / false | `true`, `false` |
| `std::string` | Text | `"available"` |

## 3. First variable program

Run this cell and inspect the output.

````c++
#include <iostream>
#include <string>

{
  int point_count = 10;
  double score = 0.85;
  bool available = true;
  std::string reason = "available";

  std::cout << "point_count = " << point_count << std::endl;
  std::cout << "score = " << score << std::endl;
  std::cout << "available = " << available << std::endl;
  std::cout << "reason = " << reason << std::endl;
}
````

Important surprise:

```cpp
true
```

prints as:

```text
1
```

and:

```cpp
false
```

prints as:

```text
0
```

By default, C++ prints booleans as `1` and `0`.

## 8. Type 4: `std::string`

`std::string` stores text.

To use `std::string`, include:

```cpp
#include <string>
```

````c++
#include <iostream>
#include <string>

{
  std::string metric_name = "ego_progress";
  std::string reason = "available";

  std::cout << "metric_name = " << metric_name << std::endl;
  std::cout << "reason = " << reason << std::endl;
}
````

In metrics code, `reason` is very useful because it tells you **why** a score is unavailable or failed.

Examples:

```cpp

std::string reason = "unavailable_empty_trajectory";
std::string reason = "collision_within_bound";

```

## 9. Changing variable values

You can change a variable after creating it.

````c++
#include <iostream>
#include <string>

{
  double score = 1.0;
  std::string reason = "available";

  std::cout << "Before violation" << std::endl;
  std::cout << "score = " << score << std::endl;
  std::cout << "reason = " << reason << std::endl;

  score = 0.0;
  reason = "collision_detected";

  std::cout << "After violation" << std::endl;
  std::cout << "score = " << score << std::endl;
  std::cout << "reason = " << reason << std::endl;
}
````

This is already close to metric logic:

```text
Start with score 1.0
If violation happens:
    score becomes 0.0
    reason explains why
```

## 10. `const`: value that should not change

Sometimes you want a value to never change.

```cpp
const double speed_limit_mps = 10.0;
```

This means:

> `speed_limit_mps` is fixed. Do not modify it later.

Why use `const`? Because it protects you from accidental changes.

````c++
#include <iostream>
#include <string>

{
  const double speed_limit_mps = 10.0;
  double ego_speed_mps = 12.0;

  double score = 1.0;
  std::string reason = "available";

  if (ego_speed_mps > speed_limit_mps) {
    score = 0.0;
    reason = "speed_too_high";
  }

  std::cout << "score = " << score << std::endl;
  std::cout << "reason = " << reason << std::endl;
}
````

## 11. Metric-like example: speed score

Rule:

```text
If ego speed is less than or equal to 10 m/s:
    score = 1.0
    reason = available
Otherwise:
    score = 0.0
    reason = speed_too_high
```

````c++
#include <iostream>
#include <string>
#include <iomanip>

{
double score = 1.0;
std::string reason = "speed_within_limit";
const double speed_limit = 10.0;
double ego_speed = 12.0;

if (ego_speed > speed_limit) {
    score = 0.0;
    reason = "speed_too_high";
}

std::cout << std::fixed << std::setprecision(1);
std::cout << "score = " << score << std::endl;
std::cout << "reason = " << reason << std::endl;
}
````

````c++
#include <iostream>
#include <string>

{
  double ego_speed_mps = 12.0;

  double score = 1.0;
  bool available = true;
  std::string reason = "available";

  if (ego_speed_mps > 10.0) {
    score = 0.0;
    reason = "speed_too_high";
  }

  std::cout << std::boolalpha;
  std::cout << "ego_speed_mps = " << ego_speed_mps << std::endl;
  std::cout << "score = " << score << std::endl;
  std::cout << "available = " << available << std::endl;
  std::cout << "reason = " << reason << std::endl;
}
````

Edit `ego_speed_mps` to `8.0` and run again. The score should become `1`.

Rule:

```text
If ttc_s < 2.0:
    score = 0.0
    reason = "ttc_too_low"
Otherwise:
    score = 1.0
    reason = "available"
```

Edit the next cell.

````c++
#include <iostream>
#include <string>

{
  const double ttc_threshold_s = 2.0;
  double ttc_s = 1.5;

  double score = 1.0;
  bool available = true;
  std::string reason = "available";

  // TODO: write the if statement here.
  if (ttc_s < ttc_threshold_s) {
    score = 0.0;
    reason = "ttc_too_low";
  }
  // Hint: if (ttc_s < ttc_threshold_s) { ... }

  std::cout << std::boolalpha;
  std::cout << "Metric: ttc_within_bound" << std::endl;
  std::cout << "ttc_s = " << ttc_s << std::endl;
  std::cout << "ttc_threshold_s = " << ttc_threshold_s << std::endl;
  std::cout << "score = " << score << std::endl;
  std::cout << "available = " << available << std::endl;
  std::cout << "reason = " << reason << std::endl;
}
````

### Solution

````c++
#include <iostream>
#include <string>

{
  const double ttc_threshold_s = 2.0;
  double ttc_s = 1.5;

  double score = 1.0;
  bool available = true;
  std::string reason = "available";

  if (ttc_s < ttc_threshold_s) {
    score = 0.0;
    reason = "ttc_too_low";
  }

  std::cout << std::boolalpha;
  std::cout << "score = " << score << std::endl;
  std::cout << "available = " << available << std::endl;
  std::cout << "reason = " << reason << std::endl;
}
````
