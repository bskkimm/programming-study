# Lecture 7 — Loops

Today’s goal:

> Process every value in a list.

Metric code often means: for each trajectory point, check something.

## 1. Range-based `for` loop

Use this when you just need each value.

````c++
#include <iostream>
#include <vector>

{
  std::vector<double> speeds{2.0, 4.0, 6.0};

  for (const auto speed : speeds) {
    std::cout << "speed = " << speed << std::endl;
  }
}
````

`const auto speed` means:

```text
For each element, call it speed.
Do not modify it.
Let C++ infer the type.
```

## 2. Index-based loop

Use this when you need the index.

````c++
#include <iostream>
#include <vector>

{
  std::vector<double> speeds{2.0, 4.0, 6.0};

  for (std::size_t i = 0; i < speeds.size(); ++i) {
    std::cout << "index = " << i << ", speed = " << speeds.at(i) << std::endl;
  }
}
````

`std::size_t` is an unsigned integer type commonly used for container indexes and sizes.

You will see it a lot in real C++ code.

## 3. Find a violation

Rule:

```text
If any speed > limit:
  score = 0
  reason = speed_too_high
```

````c++
#include <iostream>
#include <vector>
#include <string>

{
  std::vector<double> speeds{2.0, 4.0, 12.0, 6.0};
  const double speed_limit_mps = 10.0;

  double score = 1.0;
  std::string reason = "available";

  for (const auto speed_mps : speeds) {
    if (speed_mps > speed_limit_mps) {
      score = 0.0;
      reason = "speed_too_high";
      break;
    }
  }

  std::cout << "score = " << score << std::endl;
  std::cout << "reason = " << reason << std::endl;
}
````

## 4. `break`

`break` exits the loop immediately.

Useful when one violation is enough to fail the metric.

## 5. `continue`

`continue` skips the current iteration and moves to the next one.

````c++
#include <iostream>
#include <vector>

{
  std::vector<double> speeds{0.0, 2.0, 0.0, 5.0};

  for (const auto speed_mps : speeds) {
    if (speed_mps == 0.0) {
      continue;
    }

    std::cout << "moving speed = " << speed_mps << std::endl;
  }
}
````

## 6. Loop over struct points

This looks like simplified trajectory processing.

````c++
#include <iostream>
#include <vector>

struct TrajectoryPoint07
{
  double x_m{0.0};
  double y_m{0.0};
  double speed_mps{0.0};
  double time_s{0.0};
};

{
  std::vector<TrajectoryPoint07> points{
    {0.0, 0.0, 2.0, 0.0},
    {1.0, 0.0, 3.0, 0.1},
    {2.0, 0.0, 12.0, 0.2}
  };

  for (const auto & point : points) {
    std::cout << "time = " << point.time_s
              << ", x = " << point.x_m
              << ", speed = " << point.speed_mps << std::endl;
  }
}
````

Note the `&`:

```cpp
const auto & point
```

This avoids copying the whole point. We will study references next.
