# Lecture 11 — `.hpp` and `.cpp`

Today’s goal:

> Understand C++ header/source separation.

Real Autoware metrics are usually split into:

```text
metric.hpp
metric.cpp
test_metric.cpp
```

## 1. The big idea

```text
.hpp = declaration / interface
.cpp = implementation / logic
```

The header says **what exists**.

The source file says **how it works**.

## 2. Example: single-file version

Before splitting files, we might write everything in one file:

```cpp
#include <iostream>
#include <string>

struct MetricResult
{
  double score{0.0};
  bool available{false};
  std::string reason{"unavailable"};
};

MetricResult calculate_speed_metric(double speed_mps, double limit_mps)
{
  MetricResult result;
  result.available = true;
  result.score = 1.0;
  result.reason = "available";

  if (speed_mps > limit_mps) {
    result.score = 0.0;
    result.reason = "speed_too_high";
  }

  return result;
}
```

That is okay for tiny examples. Real projects split it.

## 3. Header file: `speed_metric.hpp`

```cpp
#ifndef SPEED_METRIC_HPP_
#define SPEED_METRIC_HPP_

#include <string>

struct MetricResult
{
  double score{0.0};
  bool available{false};
  std::string reason{"unavailable"};
};

MetricResult calculate_speed_metric(double speed_mps, double limit_mps);

#endif  // SPEED_METRIC_HPP_
```

The key line is the function declaration:

```cpp
MetricResult calculate_speed_metric(double speed_mps, double limit_mps);
```

It tells other files:

> This function exists. You can call it.

## 4. Source file: `speed_metric.cpp`

```cpp
#include "speed_metric.hpp"

MetricResult calculate_speed_metric(double speed_mps, double limit_mps)
{
  MetricResult result;
  result.available = true;
  result.score = 1.0;
  result.reason = "available";

  if (speed_mps > limit_mps) {
    result.score = 0.0;
    result.reason = "speed_too_high";
  }

  return result;
}
```

This file contains the implementation.

## 5. Main file: `main.cpp`

```cpp
#include <iostream>
#include "speed_metric.hpp"

int main()
{
  const auto result = calculate_speed_metric(12.0, 10.0);
  std::cout << result.reason << std::endl;
  return 0;
}
```

Because `main.cpp` includes `speed_metric.hpp`, it knows the struct and function exist.

## 6. Include guards

This part prevents the same header from being included multiple times:

```cpp
#ifndef SPEED_METRIC_HPP_
#define SPEED_METRIC_HPP_

// header content

#endif  // SPEED_METRIC_HPP_
```

You will see this style in many C++ projects.

## 7. Declaration vs definition

Declaration:

```cpp
MetricResult calculate_speed_metric(double speed_mps, double limit_mps);
```

Definition:

```cpp
MetricResult calculate_speed_metric(double speed_mps, double limit_mps)
{
  // logic
}
```

Declaration ends with `;`.

Definition has `{}` body.

````c++
#include <iostream>
#include <string>

// Notebook-only compact example.
// In real files, the struct and declaration would be in .hpp,
// and the function body would be in .cpp.

struct MetricResult11
{
  double score{0.0};
  bool available{false};
  std::string reason{"unavailable"};
};

MetricResult11 calculate_speed_metric_11(double speed_mps, double limit_mps)
{
  MetricResult11 result;
  result.available = true;
  result.score = 1.0;
  result.reason = "available";

  if (speed_mps > limit_mps) {
    result.score = 0.0;
    result.reason = "speed_too_high";
  }

  return result;
}

{
  const auto result = calculate_speed_metric_11(12.0, 10.0);
  std::cout << result.reason << std::endl;
}
````

## 8. Autoware connection

A real metric often has:

```text
src/metrics/epdms/subscores/xxx.hpp
src/metrics/epdms/subscores/xxx.cpp
test/metrics/test_xxx.cpp
```

The `.hpp` exposes the result struct and function declaration.

The `.cpp` implements the calculation.

The test calls the function and checks the result.
