# Lecture 8 — References: `const Type &`

Today’s goal:

> Understand why C++ functions often receive inputs as `const Type &`.

This is everywhere in Autoware.

## 1. Copy vs reference

When you pass a big object to a function, copying it can be expensive.

A reference lets the function use the original object.

## 2. Normal pass-by-value

This copies the input.

````c++
#include <iostream>
#include <string>

void print_reason_by_value_08(std::string reason)
{
  std::cout << "reason = " << reason << std::endl;
}

{
  std::string reason = "available";
  print_reason_by_value_08(reason);
}
````

For small examples, copying is fine. For large trajectory objects, copying is wasteful.

## 3. Pass by const reference

```cpp
const std::string & reason
```

Means:

```text
Use the original string.
Do not copy it.
Do not modify it.
```

````c++
#include <iostream>
#include <string>

void print_reason_by_const_ref_08(const std::string & reason)
{
  std::cout << "reason = " << reason << std::endl;
}

{
  std::string reason = "available";
  print_reason_by_const_ref_08(reason);
}
````

## 4. Why `const`?

Without `const`, the function could modify the input.

With `const`, the function promises:

> I only read this input.

## 5. Metric-style example

````c++
#include <iostream>
#include <vector>

struct TrajectoryPoint08
{
  double speed_mps{0.0};
};

bool has_speed_violation_08(const std::vector<TrajectoryPoint08> & points, double speed_limit_mps)
{
  for (const auto & point : points) {
    if (point.speed_mps > speed_limit_mps) {
      return true;
    }
  }
  return false;
}

{
  std::vector<TrajectoryPoint08> points{{2.0}, {12.0}, {5.0}};
  std::cout << std::boolalpha;
  std::cout << "violation = " << has_speed_violation_08(points, 10.0) << std::endl;
}
````

This function receives:

```cpp
const std::vector<TrajectoryPoint08> & points
```

Meaning:

```text
Read the vector.
Do not copy it.
Do not modify it.
```

## 6. Rule of thumb

For beginner Autoware code:

```text
Small primitive values: pass by value
  double speed_mps
  bool available
  int count

Large objects: pass by const reference
  const std::vector<T> & values
  const Trajectory & trajectory
  const std::string & name
```
