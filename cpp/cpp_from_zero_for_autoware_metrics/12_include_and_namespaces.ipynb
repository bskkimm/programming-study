# Lecture 12 — `#include` and Namespaces

Today’s goal:

> Understand how C++ brings in other code and avoids name conflicts.

## 1. Standard includes

Standard library headers use angle brackets:

```cpp
#include <iostream>
#include <string>
#include <vector>
#include <optional>
```

These come from C++ or installed libraries.

````c++
#include <iostream>
#include <string>
#include <vector>

{
  std::vector<std::string> reasons{"available", "speed_too_high"};
  std::cout << reasons.at(0) << std::endl;
}
````

## 2. Local includes

Project headers use quotes:

```cpp
#include "speed_metric.hpp"
#include "metrics/epdms/subscores/ego_progress.hpp"
```

Quotes usually mean:

> Look in this project/workspace.

## 3. What is a namespace?

A namespace is like a folder for C++ names.

Without namespace, two files may both define:

```cpp
calculate_score()
```

and conflict.

With namespace:

```cpp
namespace metrics
{
  double calculate_score()
  {
    return 1.0;
  }
}
```

The full name is:

```cpp
metrics::calculate_score()
```

````c++
#include <iostream>

namespace lecture12_metrics
{
  double calculate_score()
  {
    return 1.0;
  }
}

{
  std::cout << lecture12_metrics::calculate_score() << std::endl;
}
````

## 4. Nested namespaces

Autoware often uses long namespaces:

```cpp
namespace autoware::planning_data_analyzer::metrics
{
  // metric code here
}
```

This means the function lives inside:

```text
autoware
└── planning_data_analyzer
    └── metrics
```

````c++
#include <iostream>

namespace autoware::planning_data_analyzer::metrics::lecture12
{
  double calculate_dummy_score()
  {
    return 1.0;
  }
}

{
  std::cout << autoware::planning_data_analyzer::metrics::lecture12::calculate_dummy_score()
            << std::endl;
}
````

## 5. `using`

You can shorten long names with `using`.

Example:

```cpp
using autoware::planning_data_analyzer::metrics::calculate_score;
```

Or inside a namespace:

```cpp
using autoware::route_handler::RouteHandler;
```

Be careful: too much `using namespace ...` can make code unclear.

````c++
#include <iostream>

namespace lecture12_long_name
{
  double calculate_score()
  {
    return 0.8;
  }
}

{
  using lecture12_long_name::calculate_score;
  std::cout << calculate_score() << std::endl;
}
````

## 6. Why `std::cout` has `std::`

`cout`, `string`, and `vector` live inside the standard namespace `std`.

So we write:

```cpp
std::cout
std::string
std::vector
```

You may see:

```cpp
using namespace std;
```

But in serious C++ projects, it is often better to write `std::` explicitly.

## 7. Autoware connection

Real metric files often look like:

```cpp
#include "ego_progress.hpp"

#include "metrics/geometry/lanelet_queries.hpp"
#include "metrics/geometry/metric_utils.hpp"

#include <algorithm>
#include <memory>
#include <optional>
#include <vector>

namespace autoware::planning_data_analyzer::metrics
{

// implementation

}  // namespace autoware::planning_data_analyzer::metrics
```

Pattern:

```text
local includes first
external/standard includes next
namespace wrapper
```
