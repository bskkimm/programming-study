# Lecture 6 — `std::vector`

Today’s goal:

> Understand C++'s dynamic list container.

This is essential because Autoware trajectories contain a list of trajectory points.

## 1. What is `std::vector`?

A vector is a growable list.

Python:

```python
speeds = [1.0, 2.0, 3.0]
```

C++:

```cpp
std::vector<double> speeds{1.0, 2.0, 3.0};
```

To use it:

```cpp
#include <vector>
```

````c++
#include <iostream>
#include <vector>

{
  std::vector<double> speeds{1.0, 2.0, 3.0};

  std::cout << "size = " << speeds.size() << std::endl;
  std::cout << "first = " << speeds.front() << std::endl;
  std::cout << "last = " << speeds.back() << std::endl;
}
````

## 2. Common operations

| Operation | Meaning |
|---|---|
| `v.size()` | number of elements |
| `v.empty()` | true if no elements |
| `v.front()` | first element |
| `v.back()` | last element |
| `v.at(i)` | element at index `i` with bounds checking |
| `v.push_back(x)` | append `x` to the end |

````c++
#include <iostream>
#include <vector>

{
  std::vector<double> speeds;

  speeds.push_back(3.0);
  speeds.push_back(5.0);
  speeds.push_back(7.5);

  std::cout << "size = " << speeds.size() << std::endl;
  std::cout << "at(1) = " << speeds.at(1) << std::endl;
}
````

## 3. Index starts from zero

For:

```cpp
std::vector<double> speeds{3.0, 5.0, 7.5};
```

Indexes are:

```text
speeds.at(0) = 3.0
speeds.at(1) = 5.0
speeds.at(2) = 7.5
```

````c++
#include <iostream>
#include <vector>

{
  std::vector<double> speeds{3.0, 5.0, 7.5};

  std::cout << "0: " << speeds.at(0) << std::endl;
  std::cout << "1: " << speeds.at(1) << std::endl;
  std::cout << "2: " << speeds.at(2) << std::endl;
}
````

## 4. Check empty before `front()` / `back()`

This is dangerous:

```cpp
std::vector<double> values;
values.front();  // bad if empty
```

Always check:

```cpp
if (values.empty()) {
  // unavailable
}
```

````c++
#include <iostream>
#include <vector>
#include <string>

{
  std::vector<double> speeds;
  std::string reason = "available";

  if (speeds.empty()) {
    reason = "unavailable_empty_speeds";
  }

  std::cout << "reason = " << reason << std::endl;
}
````

## 5. Vector of structs

This prepares us for trajectory points.

````c++
#include <iostream>
#include <vector>

struct SimplePoint06
{
  double x_m{0.0};
  double y_m{0.0};
  double speed_mps{0.0};
};

{
  std::vector<SimplePoint06> points;
  points.push_back(SimplePoint06{0.0, 0.0, 2.0});
  points.push_back(SimplePoint06{1.0, 0.0, 3.0});

  std::cout << "point count = " << points.size() << std::endl;
  std::cout << "first speed = " << points.front().speed_mps << std::endl;
  std::cout << "last x = " << points.back().x_m << std::endl;
}
````

## 6. Metric-like example: short trajectory check

A trajectory with fewer than 2 points is too short for progress calculation.

````c++
#include <iostream>
#include <vector>
#include <string>

{
  std::vector<double> trajectory_x_positions{0.0};

  bool available = false;
  std::string reason = "unavailable";

  if (trajectory_x_positions.size() < 2) {
    reason = "unavailable_short_trajectory";
  } else {
    available = true;
    reason = "available";
  }

  std::cout << std::boolalpha;
  std::cout << "available = " << available << std::endl;
  std::cout << "reason = " << reason << std::endl;
}
````
